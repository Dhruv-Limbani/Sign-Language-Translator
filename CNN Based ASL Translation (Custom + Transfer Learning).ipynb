```python
# Initialize MediaPipe Hands and Drawing modules
mp_hands = mp.solutions.hands

# Set up webcam
cap = cv2.VideoCapture(0)

m = 20
# Define MediaPipe Hands model
with mp_hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.7) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Flip the image horizontally for a later selfie-view display
        frame = cv2.flip(frame, 1)
        
        # Convert the BGR image to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Process the image and detect hands
        results = hands.process(rgb_frame)
        
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                
                # Extract hand keypoints and calculate bounding box
                h, w, _ = frame.shape
                landmark_coords = [(int(lm.x * w), int(lm.y * h)) for lm in hand_landmarks.landmark]

                # Find the minimum and maximum x, y coordinates to get the hand bounding box
                min_x = min([coord[0] for coord in landmark_coords])
                max_x = max([coord[0] for coord in landmark_coords])
                min_y = min([coord[1] for coord in landmark_coords])
                max_y = max([coord[1] for coord in landmark_coords])

                # Define the fixed shape of the bounding box (e.g., 200x200 pixels)
                box_size = 200
                center_x = (min_x + max_x) // 2
                center_y = (min_y + max_y) // 2

                # Calculate the top-left corner of the fixed box
                start_x = center_x - box_size // 2
                start_y = center_y - box_size // 2

                # Calculate the bottom-right corner of the fixed box
                end_x = center_x + box_size // 2
                end_y = center_y + box_size // 2
                
                # Draw bounding box around hand
                temp_frame = frame.copy()
                cv2.rectangle(frame, (start_x-m, start_y-m), (end_x+m, end_y+m), (0, 255, 0), 2)
                
                # Crop the hand region

                cropped_hand = temp_frame[start_y-20:end_y+m, start_x-m:end_x+m]
                
                cv2.imwrite('img.jpg',cropped_hand)
        # Display the frame with bounding box and landmarks
        cv2.imshow("Hand Detection", frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Assuming 'imgs' is a list of images and 'labels' is a list of corresponding labels
# Convert lists to numpy arrays
x = np.array(imgs)
y = np.array(one_hot_labels)

# Normalize image data (assuming pixel values are in range [0, 255])
x = x.astype('float32') / 255.0


# Split data into training and validation sets (80% train, 20% validation)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
# Define a simple CNN model
def create_cnn_model(input_shape, num_classes):
    model = Sequential()

    # Convolutional layer 1
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))

    # Convolutional layer 2
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Convolutional layer 3
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Flatten the output
    model.add(Flatten())

    # Fully connected layer
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))

    # Output layer (softmax for multi-class classification)
    model.add(Dense(num_classes, activation='softmax'))

    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
# Get input shape from the first image (e.g., (height, width, channels))
input_shape = x_train.shape[1:]

# Create the CNN model
model = create_cnn_model(input_shape, 26)

# Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))

# Evaluate the model
val_loss, val_accuracy = model.evaluate(x_test, y_test)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")
```

```python
base_model.trainable = False
```

```python
model = Model(inputs=base_model.input, outputs=predictions)
```

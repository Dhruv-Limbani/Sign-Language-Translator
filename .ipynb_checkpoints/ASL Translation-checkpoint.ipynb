```python
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import time
import mediapipe as mp
```

```python
# Initialize MediaPipe Hands and Drawing modules
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
# Initialize MediaPipe Hands and Drawing modules
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

# Set up webcam
cap = cv2.VideoCapture(0)

# Define MediaPipe Hands model
with mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Flip the image horizontally for a later selfie-view display
        frame = cv2.flip(frame, 1)
        
        # Convert the BGR image to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Process the image and detect hands
        results = hands.process(rgb_frame)
        
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                # Draw landmarks and connections
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                
                # Extract hand keypoints and calculate bounding box
                h, w, _ = frame.shape
                x_min = w
                y_min = h
                x_max = y_max = 0
                
                for lm in hand_landmarks.landmark:
                    x, y = int(lm.x * w), int(lm.y * h)
                    x_min = min(x_min, x)
                    y_min = min(y_min, y)
                    x_max = max(x_max, x)
                    y_max = max(y_max, y)
                
                # Draw bounding box around hand
                cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                
                # Crop the hand region
                cropped_hand = frame[y_min:y_max, x_min:x_max]
                
        
        # Display the frame with bounding box and landmarks
        cv2.imshow("Hand Detection", frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
total_images_per_alphabets = 200
```

```python
# Define MediaPipe Hands model
cap = cv2.VideoCapture(0)

with mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7) as hands:
    
    for action in actions:
        
        cv2.waitKey(5000)
        
        for i in range(total_images_per_alphabets):
            
            ret, frame = cap.read()
            if not ret:
                break

            # Flip the image horizontally for a later selfie-view display
            frame = cv2.flip(frame, 1)

            # Convert the BGR image to RGB
            rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Process the image and detect hands
            results = hands.process(rgb_frame)
                
            cv2.putText(frame, f"Collecting frames for {action} Video Number {i}", (15,12),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
            
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

                    # Extract hand keypoints and calculate bounding box
                    h, w, _ = frame.shape
                    x_min = w
                    y_min = h
                    x_max = y_max = 0

                    for lm in hand_landmarks.landmark:
                        x, y = int(lm.x * w), int(lm.y * h)
                        x_min = min(x_min, x)
                        y_min = min(y_min, y)
                        x_max = max(x_max, x)
                        y_max = max(y_max, y)

                    # Draw bounding box around hand
                    cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

                    # Crop the hand region
                    cropped_hand = frame[y_min:y_max, x_min:x_max]
                    
                    # Extract hand keypoints into a list
                    keypoints = []
                    for lm in hand_landmarks.landmark:
                        keypoints.append([lm.x, lm.y, lm.z])

                    # Convert keypoints to numpy array
                    keypoints_array = np.array(keypoints).flatten()

                    # Save keypoints as .npy file
                    file_name = f"{i}.npy"
                    file_path = os.path.join(DATA_PATH, action, str(i))
                    np.save(file_path, keypoints_array)

            
            cv2.imshow("Hand Detection", frame)
            
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        a = input("Enter any 'c' to continue with the next action data collection")
        if(a=='c'): continue
# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
actions = [x for x in alphabets]
```

```python
actions
```

```python
label_map = {label:num for num,label in enumerate(actions)}
label_map
```

```python
imgs, labels = [], []
for action in actions:
    for i in range(50):
        res = np.load(os.path.join(DATA_PATH, action, f"{i}.npy"))
        imgs.append(res)
        labels.append(label_map[action])
```

```python
np.array(imgs).shape
```

```python
np.array(labels).shape
```

```python
x = np.array(imgs)
x.shape
```

```python
y = to_categorical(labels).astype(int)
y.shape
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)
```

```python
import tensorflow as tf

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(63,)),  # Input layer
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(26, activation='softmax')  # Output layer for 3 classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # Use this if labels are one-hot encoded
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=30, batch_size=32, validation_split=0.2)
```

```python
colors = [(245,117,16), (117,245,16), (16,117,245)]
def prob_viz(res, actions, input_frame, colors):
    output_frame = input_frame.copy()
    for num, prob in enumerate(res):
        cv2.rectangle(output_frame, (0,60+num*40), (int(prob)*100, 90+num*40), colors[num], -1)
        cv2.putText(output_frame, actions[num], (0, 85+num*40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
        
    return output_frame
```

```python
loaded_model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(63,)),  # Input layer
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(26, activation='softmax')  # Output layer for 3 classes
])

loaded_model.load_weights('asl_all.h5')
```

```python
# Initialize MediaPipe Hands
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

# Initialize webcam
cap = cv2.VideoCapture(0)

# Define the function to preprocess keypoints
def preprocess_keypoints(landmarks):
    keypoints = []
    for lm in landmarks:
        keypoints.extend([lm.x, lm.y, lm.z])
    return np.array(keypoints).flatten()

# Initialize MediaPipe Hands
with mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7) as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Flip the image horizontally
        frame = cv2.flip(frame, 1)
        h, w, _ = frame.shape

        # Convert the image to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Process the image and detect hands
        results = hands.process(rgb_frame)

        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                # Extract hand keypoints
                keypoints = preprocess_keypoints(hand_landmarks.landmark)

                # Predict the sign
                keypoints = np.expand_dims(keypoints, axis=0)  # Add batch dimension
                prediction = loaded_model.predict(keypoints)
                class_id = np.argmax(prediction)
                confidence = np.max(prediction)
                
                # Convert class_id to your class labels if needed
                class_labels = actions  # Replace with your actual labels
                class_label = class_labels[class_id]

                # Draw landmarks and bounding box
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                x_min = int(w * min(lm.x for lm in hand_landmarks.landmark))
                y_min = int(h * min(lm.y for lm in hand_landmarks.landmark))
                x_max = int(w * max(lm.x for lm in hand_landmarks.landmark))
                y_max = int(h * max(lm.y for lm in hand_landmarks.landmark))
                
                # Draw bounding box
                cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
                
                # Display the classification result
                label_text = f"{class_label} ({confidence:.2f})"
                cv2.putText(frame, label_text, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
        
        # Display the frame
        cv2.imshow('Hand Sign Classification', frame)

        # Exit on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
